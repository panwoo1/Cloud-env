강력범죄의 장소별 범죄 수

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

crime = pd.read_csv('/content/경찰청_범죄 발생 장소별 통계_20221231.csv', encoding='cp949')

# 강력범죄에 해당하는 데이터 필터링
violent_crimes = crime[crime['범죄대분류'] == '강력범죄']

# 장소(컬럼)별 범죄 수를 집계
location_crime_counts = violent_crimes.sum()[2:]

plt.figure(figsize=(10, 6))

# Seaborn을 사용하여 막대 그래프 그리기
sns.barplot(x=location_crime_counts.index, y=location_crime_counts.values, palette="Set2" ,saturation=0.7)

plt.title('강력범죄의 장소별 범죄 수')
plt.xlabel('장소')
plt.ylabel('범죄 수')
plt.xticks(rotation=90, fontsize=8)

plt.show()
```

강력범죄의 범죄별 범죄 비율 plt.pie

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
crime = pd.read_csv('/content/경찰청_범죄 발생 장소별 통계_20221231.csv', encoding='cp949')

# 강력범죄에 해당하는 데이터 필터링
violent_crimes = crime[crime['범죄대분류'] == '강력범죄']

# 범죄별로 그룹화하여 범죄별 장소별 범죄 수 계산
crime_counts = violent_crimes.groupby('범죄중분류')['아파트_연립다세대'].sum()

plt.figure(figsize=(20, 20))

# 파이 차트로 시각화
plt.pie(crime_counts, labels=crime_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('husl'))

plt.title('강력범죄의 범죄별 아파트_연립다세대 범죄 수 비율')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기
crime = pd.read_csv('/content/경찰청_범죄 발생 장소별 통계_20221231.csv', encoding='cp949')

# 강력범죄에 해당하는 데이터 필터링
violent_crimes = crime[crime['범죄대분류'] == '강력범죄']

# 장소(컬럼)별 범죄 수를 집계
location_crime_counts = violent_crimes.sum()[2:]

# 1. 범죄 유형에 따른 장소별 범죄 수 시각화
plt.figure(figsize=(15, 8))
sns.barplot(x='범죄대분류', y='아파트_연립다세대', data=crime, palette='Set2')
plt.title('범죄 유형에 따른 아파트/연립다세대별 범죄 수')
plt.xlabel('범죄 유형')
plt.ylabel('범죄 수')
plt.xticks(rotation=45)
plt.show()

# 2. 범죄 유형에 따른 범죄 발생 빈도 시각화
plt.figure(figsize=(15, 8))
sns.countplot(x='범죄대분류', data=crime, palette='Set2')
plt.title('범죄 유형에 따른 발생 빈도')
plt.xlabel('범죄 유형')
plt.ylabel('발생 빈도')
plt.xticks(rotation=45)
plt.show()

# 3. 장소별 범죄 유형 비율 시각화
plt.figure(figsize=(40, 40))
crime_location_counts = crime.iloc[:, 2:].sum()
crime_location_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('장소별 범죄 유형 비율')
plt.ylabel('')
plt.show()
```

```python
import pandas as pd
import numpy as np

# 데이터프레임 불러오기 (이미 불러온 데이터프레임을 사용하는 경우 생략 가능)
crime = pd.read_csv('/content/경찰청_범죄 발생 장소별 통계_20221231.csv', encoding='cp949')

# 특정 열 선택 ('아파트_연립다세대' 열을 예시로 선택)
selected_column = '아파트_연립다세대'

# 해당 열에 대한 평균, 중앙값, 표준편차 계산
mean_value = crime[selected_column].mean()
median_value = crime[selected_column].median()
std_deviation_value = np.std(crime[selected_column])

# 결과 출력
print(f"{selected_column}의 평균: {mean_value}")
print(f"{selected_column}의 중앙값: {median_value}")
print(f"{selected_column}의 표준편차: {std_deviation_value}")
```

```python
import pandas as pd
import numpy as np

# 데이터프레임 불러오기 (이미 불러온 데이터프레임을 사용하는 경우 생략 가능)
crime = pd.read_csv('/content/경찰청_범죄 발생 장소별 통계_20221231.csv', encoding='cp949')

# 특정 범죄 유형 선택 (다른 범죄 유형으로 변경)
selected_crime_type = '강력범죄'

# 해당 범죄 유형에 대한 평균, 중앙값, 표준편차 계산
mean_value = crime.loc[crime['범죄대분류'] == selected_crime_type].iloc[:, 2:].mean().mean()
median_value = crime.loc[crime['범죄대분류'] == selected_crime_type].iloc[:, 2:].median().median()
std_deviation_value = np.std(crime.loc[crime['범죄대분류'] == selected_crime_type].iloc[:, 2:].mean())

# 결과 출력
print(f"{selected_crime_type}의 평균: {mean_value}")
print(f"{selected_crime_type}의 중앙값: {median_value}")
print(f"{selected_crime_type}의 표준편차: {std_deviation_value}")
```
